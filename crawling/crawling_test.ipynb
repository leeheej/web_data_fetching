````python
response = requests.get("http://leeeheee.com")
rating_page = response.text
print(rating_page)
````

````python
# rating_page 안에 있는 html 코드를 html.parser를 통해서 파싱
soup = BeautifulSoup(rating_page, 'html.parser')
print(soup.prettify())  # html 코드를 예쁘게 출력
````

````python
print(soup.select('title'))
print('-----')
# soup 변수는 html.parser를 통해서 파싱된 html 코드를 담고 있음
# soup 변수 안에 있는 html 코드 중에서 특정 태그를 찾아보자
# 예를 들어, <title> 태그를 찾아보자
title_tag = soup.title
print(title_tag)  
print(title_tag.string) 
print(title_tag.name)  
print(title_tag.attrs) 
````

````python
# css 선택자로 html 불러오기
soup.select('h1.logo')
````

````python
soup.select('div.profile > p')
````

````python
# for 문으로 text 순차 출력하기
for profile in soup.select('div.profile > p'):
    print(profile.text)
````

````python
# list로 출력하기
profile_list = []

for p in soup.select('div.profile > p'):
    profile_list.append(p.text)
    
print(profile_list)
````

````python
for p in soup.select('div.profile > p'):
    print(list(p.strings)) # strings 는 접근만 가능하기 때문에 list()로 감싸서 리스트로 변환
````

````python
for p in soup.select('div.profile > p'):
    print(list(p.stripped_strings)[0]) # 앞 뒤 공백 제거 + 0번 인덱스 접근
````

>
Beautiful Soup에는 필요한 태그를 선택하는 방법이 크게 두 가지 있습니다:
```
.select_one(), .select() 메소드 활용
.find(), .find_all() 메소드 활용
```
1번, 2번 다 기능이 비슷하기 때문에 둘 중 하나만 아셔도 무방한데요, 2번 방법도 많이 쓰이기 때문에 이번 레슨에서 설명드리겠습니다.

### find 와 find_all
find()는 select_one()과 비슷하고, find_all()은 select()와 비슷합니다.
```
find()는 매칭되는 태그를 (최대) 하나만 가져오는데 쓰이고 
find_all()은 매칭되는 태그를 모두 가져오는데 쓰이는 거죠. 
리턴값의 형식도 똑같습니다. 
find()는 태그 자체나 None을 리턴하고 find_all()은 리스트를 리턴합니다.

하지만find(), find_all()은 CSS 선택자를 쓰지 않고 여러 파라미터를 활용합니다. 
```

````python
soup.find_all(['p', 'a']) # p 태그와 a 태그를 모두 찾아서 리스트로 리턴

# 모든 태그 가져오기
# soup.find_all(True)
````

````python
# soup.find_all('tagname', attr1='val1', attr2='val2', ...)
soup.find_all('h5', id_="graphic")
````

````python
# select - 선택자로 중첩된 p 태그 선택
soup.select('div.some-class p')

# find_all - 먼저 div 태그를 가져온 후 div 태그에 find_all()
div_tag = soup.find('div', class_='some-class')
div_tag.find_all('p')
````

````python
# img 태그 가져오기
print(soup.select_one('img')['src'])
print(soup.select_one('img').attrs)
````
